```python
with open("RussellTickers500to1000.csv") as f:
    tickers = f.read()

dftot=yf.download(tickers, start='2012-02-01', end='2022-06-01', interval = "1mo", group_by="ticker")

dftot = dftot.xs('Close', level=1, axis=1)
dftot = dftot.reset_index()
dftot = dftot.fillna(0)

dftot.to_csv('RussellData500to1000.csv', index=False)

dftot.head()
```
