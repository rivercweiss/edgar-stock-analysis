```python
import yfinance as yf
import numpy as np
import csv
import pandas as pd
```

```python
dfall=pd.read_csv('10yrtickers.csv')
dfall.head()
```

```python
df5yr = dfall[(dfall["Date"] == '2017-05-01') | (dfall["Date"] =='2022-05-01')]
df5yr = df5yr.bfill().ffill()[::df5yr.shape[0]-1]
df5yr=df5yr.drop(columns='Date')
df5yr=df5yr.reset_index(drop=True)
df5yr=df5yr.T
df5yr["diff"]=(df5yr[1]/df5yr[0])
df5yr["5 Year CAGR"]= (df5yr["diff"].pow(1./5)-1)*100
df5yr=df5yr[["5 Year CAGR"]]
df5yr=df5yr.dropna()
df5yr = df5yr[np.isfinite(df5yr).all(1)]
df5yr.head(20)
```

```python
df2yr=dfall[(dfall["Date"] == '2020-05-01') | (dfall["Date"] =='2022-05-01')]
df2yr = df2yr.bfill().ffill()[::df2yr.shape[0]-1]
df2yr=df2yr.drop(columns='Date')
df2yr=df2yr.reset_index(drop=True)
df2yr=df2yr.T
df2yr["diff"]=(df2yr[1]/df2yr[0])
df2yr["2 Year CAGR"]= (df2yr["diff"].pow(1./2)-1)*100
df2yr=df2yr[["2 Year CAGR"]]
df2yr=df2yr.dropna()
df2yr = df2yr[np.isfinite(df2yr).all(1)]
df2yr.head(20)
```

```python
df10yr=dfall[(dfall["Date"] == '2012-05-01') | (dfall["Date"] =='2022-05-01')]
df10yr = df10yr.bfill().ffill()[::df10yr.shape[0]-1]
df10yr=df10yr.drop(columns='Date')
df10yr=df10yr.reset_index(drop=True)
df10yr=df10yr.T
df10yr["diff"]=(df10yr[1]/df10yr[0])
df10yr["10 Year CAGR"]= (df10yr["diff"].pow(1./10)-1)*100
df10yr=df10yr[["10 Year CAGR"]]
df10yr=df10yr.dropna()
df10yr = df10yr[np.isfinite(df10yr).all(1)]
df10yr.head(20)
```

```python
df=df2yr.join(df5yr)
df=df.join(df10yr)
df["Average CAGR 2 & 5"]=(df["2 Year CAGR"]+df["5 Year CAGR"])/2
df["Average CAGR All"]=(df["2 Year CAGR"]+df["5 Year CAGR"]+df['10 Year CAGR'])/3
df=df.reset_index()
df.rename(columns={'index' :'Symbol'}, inplace=True )
df.head()
```

```python
df2=pd.read_csv('/Users/river/Documents/Code/Yfinance Code/constituents.csv')
df= pd.merge(df, df2, on="Symbol")
df=df.sort_values(by="Average CAGR 2 & 5", ascending=False)
df.head(50)
```

```python
df=df.sort_values(by="Average CAGR All", ascending=False)
df.head(50)
```

```python
dfs=df[df["10 Year CAGR"]>15].head(100).sort_values(by="10 Year CAGR", ascending=False)
dfs.head(100).to_csv("100 S&P 500 Top Steady Returning Companies.csv", index=False)
dfs.head(50)
```
